<p style="font-size:20px;">
<strong><span style="color:#7C3AED;">Columns : 12</span></strong>
</p>

<p><strong><span style="color:#DC2626;">1. ID</span></strong> - it is just IDs of the orders which will not be good for our training data so we will <mark>drop it</mark>.</p>

<p><strong><span style="color:#16A34A;">2. Warehouse_block</span></strong> - it represents only <mark><strong>5 unique values</strong></mark> and we will keep this column.</p>

<p><strong><span style="color:#16A34A;">3. Mode_of_Shipment</span></strong> - it will play a integral role for our model to predict shipping delay. It represents the mode of shipment and have <mark><strong>3 unique values</strong></mark>.</p>

<p><strong><span style="color:#16A34A;">4. Customer_care_calls</span></strong> - it represents the number of customer care calls made by the customer and have <mark><strong>6 unique values</strong></mark>. This feature can have a relationship with shipping delay so we will keep this column.</p>

<p><strong><span style="color:#DC2626;">5. Customer_rating</span></strong> - it represents the customer rating and have <mark><strong>5 unique values</strong></mark>. Since the rating is generally given after the delivery process it can introduce <mark><strong>data leakage</strong></mark> so we will drop this column.</p>

<p><strong><span style="color:#16A34A;">6. Cost_of_the_Product</span></strong> - it represents the cost of the product and have <mark><strong>215 unique values</strong></mark>. Product cost may affect the shipping priority so we will keep this column.</p>

<p><strong><span style="color:#16A34A;">7. Prior_purchases</span></strong> - it represents the number of previous purchases made by the customer and have <mark><strong>8 unique values</strong></mark>. This feature can help our model identify customer purchasing behaviour so we will keep this column.</p>

<p><strong><span style="color:#16A34A;">8. Product_importance</span></strong> - it represents the importance level of the product and have <mark><strong>3 unique values</strong></mark>. Product importance can influence shipping priority so we will keep this column.</p>

<p><strong><span style="color:#DC2626;">9. Gender</span></strong> - it represents the gender of the customer and have <mark><strong>2 unique values</strong></mark>. This feature has no meaningful relationship with shipping delay so we will drop this column.</p>

<p><strong><span style="color:#16A34A;">10. Discount_offered</span></strong> - it represents the discount offered on the product and have <mark><strong>65 unique values</strong></mark>. This feature can influence shipping behaviour so we will keep this column.</p>

<p><strong><span style="color:#16A34A;">11. Weight_in_gms</span></strong> - it represents the weight of the product in grams and have <mark><strong>4034 unique values</strong></mark>. Product weight can directly affect shipping time so we will keep this column.</p>

<p><strong><span style="color:#2563EB;">12. Reached.on.Time_Y.N</span></strong> - it represents whether the shipment reached on time or not and have <mark><strong>2 unique values</strong></mark>. This is our <mark><strong>target variable (y)</strong></mark>.</p>


```python
#encoding

#setting up multiple encoding models for future streamlit app

from sklearn.preprocessing import LabelEncoder

warehouse_encoder = LabelEncoder()
shipment_encoder = LabelEncoder()
importance_encoder = LabelEncoder()

df['Warehouse_block'] = warehouse_encoder.fit_transform(df['Warehouse_block'])
df['Mode_of_Shipment'] = shipment_encoder.fit_transform(df['Mode_of_Shipment'])
df['Product_importance'] = importance_encoder.fit_transform(df['Product_importance'])
```

```python
df.columns
```

```python
#X and y

X = df.drop(columns=["ID", "Gender", "Reached.on.Time_Y.N",'Customer_rating'])

y = df["Reached.on.Time_Y.N"]
```

```python
df_encoded = df.copy()
```

```python
# Saving the encoded dataset as a csv file

save_path = r"C:\Users\shobh\Downloads\AI and DATA SCIENCE\𝗠𝗟 𝗣𝗿𝗼𝗷𝗲𝗰𝘁𝘀\Shipping Delay Prediction System\data\encoded.csv"

df_encoded.to_csv(save_path, index=False)
```

```python
#after ncoding next step step is train_test_split


from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=42)

X_train.head()
```

```python
#now encodind and train test split done now we will perform feature scaling


from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
# now first we will start with models that usually  need scaling 

# 1. Logistic Regression
# 2. k-nearest neighbors
# 3. SVM


# we will train each model note their evaluation metrics and compare all models together at last and perform model tuning on best 3 models 
```

```python
#logistic regression 


from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

lr_model = LogisticRegression(random_state = 42 , max_iter = 10000)


lr_model.fit(X_train_scaled,y_train)

y_pred_lr = lr_model.predict(X_test_scaled)
print("Accuracy lr:", accuracy_score(y_test, y_pred_lr))
print("Precision lr:", precision_score(y_test, y_pred_lr))
print("Recall   lr :", recall_score(y_test, y_pred_lr))
print("F1 Score lr :", f1_score(y_test, y_pred_lr))
```

```python
# same with knn now

from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5)

knn_model.fit(X_train_scaled, y_train)

y_pred_knn = knn_model.predict(X_test_scaled)

print("Accuracy knn:", accuracy_score(y_test, y_pred_knn))
print("Precision knn:", precision_score(y_test, y_pred_knn))
print("Recall   knn :", recall_score(y_test, y_pred_knn))
print("F1 Score knn :", f1_score(y_test, y_pred_knn))
```

</p>

```python
#StackingClassifier (randomsearchcv) 


from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# Base models
estimators = [
    ("rf", RandomForestClassifier(random_state=42)),
    ("dt", DecisionTreeClassifier(random_state=42))
]

# Stacking model
stack_model = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(max_iter=10000, random_state=42)
)

param_dist = {
    "rf__n_estimators": randint(50, 300),
    "rf__max_depth": [None, 5, 10, 20],
    "dt__max_depth": [None, 5, 10, 20],
    "final_estimator__C": [0.01, 0.1, 1, 10, 100]
}

random_search = RandomizedSearchCV(
    estimator=stack_model,
    param_distributions=param_dist,
    n_iter=20,
    scoring="f1",
    cv=5,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train, y_train)

print("Best Parameters:", random_search.best_params_)
print("Best F1 Score:", random_search.best_score_)
```

```python
# Evaluate RandomizedSearchCV Model

best_stack_random = random_search.best_estimator_

y_pred_random = best_stack_random.predict(X_test)

print("Accuracy :", accuracy_score(y_test, y_pred_random))
print("Precision:", precision_score(y_test, y_pred_random))
print("Recall   :", recall_score(y_test, y_pred_random))
print("F1 Score :", f1_score(y_test, y_pred_random))
```

```python
# Stacking Classifier ( GridSearchCV)

from sklearn.model_selection import GridSearchCV

param_grid = {
    "rf__n_estimators": [150, 200],
    "rf__max_depth": [10, None],
    "dt__max_depth": [5, None],
    "final_estimator__C": [1, 10]
}

grid_search = GridSearchCV(
    estimator=stack_model,
    param_grid=param_grid,
    scoring="f1",
    cv=3,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best F1 Score:", grid_search.best_score_)
```
